## **Parametros de la funcion read_csv**




```python
# Leemos el archivo usando sep=';'
datos_punto_coma = pd.read_csv('superstore_data_punto_coma.csv', sep=';')
```

```python
datos_punto_coma.head()
```

```python
# nrows nos muestra la cantidad de lineas que le indicamos
datos_primeras_lineas = pd.read_csv('superstore_data.csv', nrows= 5)
datos_primeras_lineas
```

```python
# usecols me sirve para conocer la informacion de algunas columnas en especial
datos_seleccion = pd.read_csv('superstore_data.csv', usecols=['Id','Year_Birth','Income'])
datos_seleccion
```

```python
# Con usecols tambien podemos ver columnas por medio de la posicion en la que se encuentran
datos_seleccion = pd.read_csv('superstore_data.csv', usecols=[0,1,4])
datos_seleccion
```

## **Escribiendo archivos CSV**

## **Leyendo datos de Google Sheets**

```python
link: 'https://docs.google.com/spreadsheets/d/1_hVzrIsa2cX4wriDz6HUXImxCeVq1YxxLo9kfdnZ0JM/edit?usp=sharing'
```

```python
# Sacamos el id de la hoja de Google sheet que esta dentro del link anterior.
# Luego ponemos el link en una variable y le cambiamos el id por la variable sheet_id en
# f string, para que no sea tan largo.
# Corto edit... y lo que sigue porque es para visualizar en g.sheets, para comunicarse con
# google colab debo poner gviz/tq.
# Debo agregar el formato de salida para google colab o sea en csv. Se coloca tqx=out:csv
# sheet se pone en el link para especificar que quiero traer una hoja de g. sheet

sheet_id = '1_hVzrIsa2cX4wriDz6HUXImxCeVq1YxxLo9kfdnZ0JM'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
# Leemos el archivo
datos_co2_sheets = pd.read_csv(url)
```

```python
# Nos abre la primera hoja.
datos_co2_sheets
```

```python
# Para abrir otras hojas.
# En la url despues de sheet coloco el nombre de la variable que cree con el nombre de la hoja.
sheet_id = '1_hVzrIsa2cX4wriDz6HUXImxCeVq1YxxLo9kfdnZ0JM'
sheet_name = 'emisiones_percapita'
url_percapita = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```

```python
percapita_sheets = pd.read_csv(url_percapita)
```

```python
percapita_sheets
```

```python
# Abrimos la tercera hoja
sheet_id = '1_hVzrIsa2cX4wriDz6HUXImxCeVq1YxxLo9kfdnZ0JM'
sheet_name = 'fuentes'
url_fuentes = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```
