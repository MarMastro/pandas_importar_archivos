# **Leyendo paginas en HTML y XML**

## **Importando paginas web**

```python
import pandas as pd
```

```python
datos_html = pd.read_html('peliculas_wikipedia.html')
```

```python
# El df no se ve correctamente porque hay tres tablas en el mismo df.
datos_html
```

```python
# Vemos que tipo de dato es.
type(datos_html)
```

```python
# Para saber el tamanio. Una lista de tres tablas.
len(datos_html)
```

```python
# Nos interesa leer los datos de la segunda tabla.
top_peliculas = datos_html[1]
```

## **Escribiendo archivos HTML**

```python
# Me devuelve una tabla.
top_peliculas.to_html('top_peliculas.html', index = False)
```

```python
# Leemos el archivo.
pd.read_html('top_peliculas.html')
```
