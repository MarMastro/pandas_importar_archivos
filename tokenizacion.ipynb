```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

```python
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

```python
columnas = list(datos.columns)
columnas
```

```python
# Separamos los datos que aparecen en columnas anidadas.
datos = datos.explode(columnas[3:])
datos
```

```python
# Reseteamos el index para que cada fila sea unica y el index no tenga varios indices repetidos.
datos.reset_index(inplace = True, drop = True)
datos.head()
```

```python
datos.info()
```

```python
# Vamos a convertir los object para numeros enteros.
import numpy as np
```

```python
# Transformamos la columna max_huespedes de object a enteros.
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
# Vemos que la columna max_hospedes cambio a int64.
datos.info()
```

```python
# Creo una lista con las columnas a convertir a enteros, para que sea
# mas sencilla la conversion.
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
# Convierto col_numericas a enteros.
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
# Vemos que en cuatro columnas ya las cambio de object a int.
datos.info()
```

```python
# Convertimos la columna evaluacion_general a float64.
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
# Convirtio la columna evaluacion_general a float64.
datos.info()
```

```python
# Vamos a quitar los caracteres que no sean numeros a las columnas ($, o comas y puntos.)
# Usamos apply para aplicar lo que queremos a todas los datos
# Aplicamos la funcion lambda a los valores x. x son los valores de la columna precio.
# Reemplazamos $ por vacio y , por vacio.
# Strip() eliminamos los espacios adelante, atras o entremedio que quede al reemplazar los
# caracteres
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
# Transformando las columnas cuota_deposito y cuota_limpieza al mismo tiempo.
# Observamos que tienen las columnas.
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
# Cuando trabajamos con una sola columna se llama series.
# Con dos o mas columnas se llama df y no podemos usar apply. Hay que usar applymap
datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
# Cambiamos los caracteres por espacios vacios y retiramos los espacios vacios con strip()
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

# **Tratar datos de textos**

```python
datos.head()
```

```python
# Debemos tratar los strings porque hay mayusculas, minusculas, caraceteres
# especiales, etc
datos['descripcion_local']
```

```python
# Transformamos la columna a string porque esta en formato object.
# Despues lowercase.
datos['descripcion_local'].str.lower()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
# Eliminamos los caracteres especiales con Regex (,.*$ etc)
# Mostramos un registro para ver los caracteres que estan en un registro como muestra.
datos['descripcion_local'][3169]
```

```python
# Regex
# Regex va siempre entre corchetes.
# ^ significa que esto no lo reemplace.
# [^a-zA-Z0-9\-\'] Significa no reemplaces de la a a la z, de la A a la Z y del 0 al 9
# La barra sirve para que no reemplace el -. Si solo ponemos el guion interpreta que
# queremos separar algo. La segunda barra es para decirle que reemplace la '
datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

```python
# Reemplazamos los guiones que estan entre dos palabras. Ej. one - head
# pero no las palabras compuestas en ingles como business-themed.
# Para eso en el regex le pasamos que quite los guiones que no estan pegados a
# un caracter a la izquierda y a la derecha.
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?<!\w)', ' ', regex = True)
```

```python
datos['descripcion_local'].head()
```

# **Tokenizacion de strings**

```python
# Creamos una lista de palabaras. Ej. Voy al cine. [Voy, al, cine]
# Las palabras las separamos por medio de un espacio como esta en el original,
# aca lo hacemos con split(). Al no poner nada dentro de los parentesis por default
# separa por espacio.
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
# En la columna comodidades voy a reemplazar con regex.
# '\{|}|\"' significa busca { y } y "
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"', '', regex = True)
```

```python
datos.head()
```

```python
# Creamos una lista de elementos. El separador sera una , como en el original
datos['comodidades'] = datos['comodidades'].str.split(',')
```
