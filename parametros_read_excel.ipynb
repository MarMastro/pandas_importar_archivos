# **Leyendo archivos Excel**


## **Importando archivos Excel**

```python
import pandas as pd
```

```python
# Guardo el archivo Excel en una variable.
archivo = 'emisiones_CO2.xlsx'
```

```python
# Leemos el archivo Excel
datos_c02 = pd.read_excel(archivo)
datos_c02
```

```python
# Metodo para leer las hojas del Excel
pd.ExcelFile(archivo).sheet_names
```

## **Parametros de la funcion read_excel**

```python
# Metodo para acceder a otras hojas del archivo. sheet_name('nombre de la hoja')
percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
```

```python
percapita.tail()
```

```python
# Metodo para leer cinco registros aleatorios
percapita.sample(5)
```

```python
# Revisamos la ultima hoja
fuentes = pd.read_excel(archivo, sheet_name='fuentes')
```

```python
fuentes.tail()
```

```python
# Selecciono las primeras cuatro columnas con usecols
intervalos = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D')
```

```python
intervalos
```

```python
# Selecciono las diez primeras filas para visualizar
intervalos2 = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D', nrows = 10)
```
